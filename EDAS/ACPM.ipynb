```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_acpm_consolidado = pd.read_csv("Archivos_Consolidados/df_acpm_230523.csv", sep=",")
df_acpm_consolidado = df_acpm_consolidado.iloc[:, 1:]
```

```python
df_acpm_consolidado.head(2)
```

Grafico 1 - GALONES ACPM COMPRADOS

```python
import matplotlib.pyplot as plt

# Obtener los meses únicos en el orden en que aparecen en el dataframe
orden_meses = df_acpm_consolidado['Mes'].unique()

# Agrupar por mes y sumar los galones
df_agrupado = df_acpm_consolidado.groupby('Mes')['Cantidad_Galones'].sum().reindex(orden_meses)

# Crear la figura y el eje
fig, ax = plt.subplots()

# Configurar los colores
color = '#8B6914'  # Color dorado oscuro
label_color = 'black'  # Color de las etiquetas

# Crear el gráfico de barras
barplot = df_agrupado.plot(kind='bar', color=color, ax=ax)

# Configurar etiquetas y título
ax.set_xlabel('Mes')
ax.set_ylabel('Cantidad de Galones')
ax.set_title('Suma de Galones por Mes')

# Añadir formato dorado oscuro
ax.spines['bottom'].set_color(label_color)
ax.spines['top'].set_color(label_color)
ax.spines['left'].set_color(label_color)
ax.spines['right'].set_color(label_color)
ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)
ax.title.set_color(label_color)
ax.xaxis.label.set_color(label_color)
ax.yaxis.label.set_color(label_color)

# Mostrar el valor de cada barra encima de ella
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', color=label_color)

# Mostrar el gráfico
plt.show()
```

**Grafico 2 - COSTO GALONES ACPM COMPRADOS**

```python
import matplotlib.pyplot as plt

# Obtener los meses únicos en el orden en que aparecen en el dataframe
orden_meses = df_acpm_consolidado['Mes'].unique()

# Agrupar por mes y sumar los galones
df_agrupado = df_acpm_consolidado.groupby('Mes')['Costo_Total_ACPM'].sum().reindex(orden_meses)

# Convertir el costo total del ACPM a millones
df_agrupado = df_agrupado / 1000000

# Crear la figura y el eje
fig, ax = plt.subplots()

# Configurar los colores
color = '#8B6914'  # Color dorado oscuro
label_color = 'black'  # Color de las etiquetas

# Crear el gráfico de barras
barplot = df_agrupado.plot(kind='bar', color=color, ax=ax)

# Configurar etiquetas y título
ax.set_xlabel('Mes')
ax.set_ylabel('Costo ACPM (millones)')
ax.set_title('Costo Total ACPM en millones')

# Añadir formato dorado oscuro
ax.spines['bottom'].set_color(label_color)
ax.spines['top'].set_color(label_color)
ax.spines['left'].set_color(label_color)
ax.spines['right'].set_color(label_color)
ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)
ax.title.set_color(label_color)
ax.xaxis.label.set_color(label_color)
ax.yaxis.label.set_color(label_color)

# Mostrar el valor de cada barra encima de ella
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.2f}M', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', color=label_color)

# Mostrar el gráfico
plt.show()
```
